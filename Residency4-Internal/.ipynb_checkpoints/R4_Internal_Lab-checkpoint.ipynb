## K-Nearest-Neighbors

KNN falls in the supervised learning family of algorithms. Informally, this means that we are given a labelled dataset consiting of training observations (x,y) and would like to capture the relationship between x and y. More formally, our goal is to learn a function h:X→Y so that given an unseen observation x, h(x) can confidently predict the corresponding output y.

In this module we will explore the inner workings of KNN, choosing the optimal K values and using KNN from scikit-learn.

## Question 2 - Estimating missing values

*Its not good to remove the records having missing values all the time. We may end up loosing some data points. So, we will have to see how to replace those missing values with some estimated values (median) *

```python
## Calculate median for not null rows for each of the columns and replace in the dataframe
## Verify the same by running count and verifying the corresponding distance

for column in iris_df.columns:
    if(column != 'Species'):
        print('the median value for not nan values in column ',column,' is ',iris_df[np.logical_not(iris_df[column].isnull())][column].median())
        iris_df[column].fillna(iris_df[np.logical_not(iris_df[column].isnull())][column].median(), inplace=True)
        
```

## Question 7

*Split the dataset into training and test sets with 80-20 ratio.*

```python
from sklearn.model_selection import train_test_split
from scipy.stats import zscore
X = iris_df[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
Y = iris_df['TSpecies']
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=100)
```

## Question 8 - Model

*Build the model and train and test on training and test sets respectively using **scikit-learn**. Print the Accuracy of the model with different values of **k=3,5,9**.*

**Hint:** For accuracy you can check **accuracy_score()** in scikit-learn

```python
## Though scaling has to be performed for KNN as it is distance based, all the independant variables are scaled as cms
## hence, explicit scaling is not required
## and the model can be used directly

from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

k = {3,5,9}
k_value = []
accuracy_score_series = []

for value in k:
    KNN = KNeighborsClassifier(n_neighbors= value,metric='euclidean',algorithm='ball_tree')
    KNN.fit(X_train,Y_train)
    accuracy_score_series.append(KNN.score(X_test,Y_test))
    k_value.append(value)
    ##print('accuracy score for k value - ',value,' is ',KNN.score(X_test,Y_test))
    Y_pred = KNN.predict(X_test)
    Y_pred_pro = KNN.predict_proba(X_test)
    ##print(accuracy_score(Y_pred,Y_test))
data = {'k-value':k_value,'accuracy-score':accuracy_score_series}
acc_score_df = pd.DataFrame(data=data)
acc_score_df
```

## Question 9 - Cross Validation

Run the KNN with no of neighbours to be 1,3,5..19 and *Find the **optimal number of neighbours** from the above list using the Mis classification error

Hint:

Misclassification error (MSE) = 1 - Test accuracy score. Calculated MSE for each model with neighbours = 1,3,5...19 and find the model with lowest MSE

```python
from sklearn.model_selection import cross_val_score
acc_cv=[]
k_cv=[]
k = np.arange(1,20,2)
for value in k:
    KNN_CV = KNeighborsClassifier(n_neighbors= value)
    scores = cross_val_score(KNN_CV,X_train,Y_train,cv=10,scoring='accuracy')
    acc_cv.append(scores.mean())
    k_cv.append(value)

MSE = [1-x for x in acc_cv]
d_cv = {'k-value':k_cv,'acc_cv':acc_cv,'MSE':MSE}
d_cv_df = pd.DataFrame(data=d_cv)
print('The minimum misclassification error is ',np.min(d_cv_df['MSE']))
print('The k-value with minimum misclassification error is ' ,k_cv[MSE.index(min(MSE))])
```
