<h3>LogisticRegressionWithCostFunction</h3>

```
def sigmoid(key):
  #return np.reciprocal(1+np.exp(-key))
  return .5 * (1 + np.tanh(.5 * key))
```

```
X_t = np.array(X_train, dtype=np.float128)
Y_t = np.array(Y_train, dtype=np.float128)
X_t1 = np.array(X_test, dtype=np.float128)
Y_t1 = np.array(Y_test, dtype=np.float128)
print(X_t.shape,Y_t.shape,X_t1.shape,Y_t1.shape)
```

```
def lossfn(X_train, Y_train, iterations, learningRate):
  W0 = np.zeros(X_train.shape[0])
  W = W0.reshape(W0.shape[0],1)
  B = 0  
  for i in np.arange(0,iterations):
    Z = np.dot(W.T,X_train)+B
    A = np.reciprocal(1+np.exp(-Z))
    print(Z)
    print(A)
    print(Y_train)
    L = - (Y_train * np.log(A) + (1-Y_train) * np.log(1-A))
    m = X_train.shape[1]
    dW = 1/m * np.dot(X_train, (A-Y_train).T)
    cost = np.squeeze(1/m * np.sum(L))
    if(i%100==0):
      print('Cost function for iteration', i, 'is ',cost)
    db = 1/m * np.sum(A-Y_train)
    lr = learningRate
    W = W - (lr * dW)
    B = B - (lr * db)
  return W, B, cost
```
