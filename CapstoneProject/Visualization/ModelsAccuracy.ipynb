```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import pandas as pd
```

> # Model_Accuracy_V2
Version V2 is got by freezing the convolutional layers and unfreezing the fully connected layers

```python
models=['vg16','vg19','res','dens','inc','incres','xcep','mob','mobv2','nnetl','nnetm']
train_btch_acc = [93.93,100.0,99.37,99.58,90.42,90.42,99.17,98.12,96.46,98.40,94.79]
val_btch_acc =   [94.03,94.03,26.86,82.09,79.10,79.60,81.09,84.57,82.09,81.01,83.58]
test_btch_acc =  [97.71,92.92,33.33,75.75,74.74,67.67,82.82,68.68,76.76,86.86,74.74]
test_acc =       [87.27,86.26,29.29,75.15,69.49,00.00,74.94,82.42,74.54,00.00,00.00]
```

```python
model_acc_v2 = pd.DataFrame({'Models_V2':models,'Train_Batch_Accuracy':train_btch_acc,'Validation_Batch_Accuracy':val_btch_acc,'Test_Batch_Accuracy':test_btch_acc, 'Test_Accuracy':test_acc})
model_acc_v2
```

```python
model_acc_v2.to_csv('Model_Acc_V2.csv', header=True, index=False)
```

```python
sns.barplot('Models_V2', 'Train_Batch_Accuracy', data=model_acc_v2).set_title("Train_Batch_Accuracy for Models_V2")
```

```python
sns.barplot('Models_V2','Validation_Batch_Accuracy', data=model_acc_v2).set_title("Validation_Batch_Accuracy for Models_V2")
```

```python
sns.barplot('Models_V2', 'Test_Batch_Accuracy', data=model_acc_v2).set_title("Test_Batch_Accuracy for Models_V2")
```

```python
sns.barplot('Models_V2', 'Test_Accuracy', data=model_acc_v2).set_title("Test_Accuracy for Models_V2")
```

```python
plt.title('TrainValTestAccForV2Models')
plt.plot(model_acc_v2['Models_V2'], model_acc_v2['Train_Batch_Accuracy'])
plt.plot(model_acc_v2['Models_V2'], model_acc_v2['Validation_Batch_Accuracy'])
plt.plot(model_acc_v2['Models_V2'], model_acc_v2['Test_Batch_Accuracy'])
plt.plot(model_acc_v2['Models_V2'], model_acc_v2['Test_Accuracy'])
plt.xlabel('Models_V2')
plt.ylabel('Accuracy')
plt.legend(['train', 'val','test','overall-test'], loc='lower left')
plt.show()
```
